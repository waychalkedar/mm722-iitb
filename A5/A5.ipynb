A5 involves estimating the self-diffusivity/diffusion constant of water through multiple methods in a simulation with varying runtimes. <br>
**Please run the input script to generate the appropriate output files to test out the code yourself.** You can take only the smallest simulation time (200.0 ps) or even lesser by varying the `runtime` parameter in the input script.

`msd_lmp.py` plots the MSD data outputted from LAMMPS vs time, along with a linear fit for the data to find the slope. The slope is then used to estimate the diffusion constant. <br>
`msd_calc.py` plots the MSD data as calculated by me vs time, along with a linear fit for the data to find the slope. Looking through LAMMPS documentation on how the `compute msd` command works, I attempted to implement a center-of-mass (COM) correction and a running average to get results closest to the MSD data generated by LAMMPS. <br>
`vacf.py` plots the time curve for the velocity auto-correlation function (VACF). This is used to calculate the diffusion constant through a Green-Kubo relation (detailed more in `A5_submission.pdf`). <br>
`nve_nvt.py` compares the MSD (LAMMPS) vs time and VACF vs time curves for different simulation runtimes, and their corresponding diffusion constant values. <br>
`press_temp.py` outputs the time variation of the system's pressure and temperature for different simulation runtimes.
